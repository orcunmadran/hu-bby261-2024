```python
import pandas as pd
import matplotlib.pyplot as plt

veriSeti = pd.read_csv('olympics.csv')
ulkeler = veriSeti['country']
altin_madalyalar = veriSeti['gold']

plt.figure(figsize=(20, 7))  # Grafiğin boyutunu ayarlayın
plt.bar(ulkeler, altin_madalyalar)
plt.xlabel('Ülkeler')
plt.ylabel('Altın Madalya Sayısı')
plt.title('Olimpiyatlarda Ülkelere Göre Altın Madalya Sayıları')
plt.xticks(rotation=90)  # Ülke isimlerinin dikey yazılması
plt.show()
```

```python
nufus = veriSeti['gold'] / veriSeti['population']

plt.figure(figsize=(20, 7))  # Grafiğin boyutunu ayarlayın
plt.bar(ulkeler, nufus)
plt.xlabel('Ülkeler')
plt.ylabel('Madalya Sayısının Nüfusa Oranı')
plt.title('Olimpiyatlar Alınan Altın Madalyaların Nüfusa Oranı')
plt.xticks(rotation=90)  # Ülke isimlerinin dikey yazılması
plt.show()
```

```python
gelir = veriSeti['gold'] / veriSeti['gdp']

plt.figure(figsize=(20, 7))  # Grafiğin boyutunu ayarlayın
plt.bar(ulkeler, gelir)
plt.xlabel('Ülkeler')
plt.ylabel('Madalya Sayısının Nüfusa Oranı')
plt.title('Olimpiyatlar Alınan Altın Madalyaların Gelire Oranı')
plt.xticks(rotation=90)  # Ülke isimlerinin dikey yazılması
plt.show()
```
