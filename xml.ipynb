**Python XML Hackathon**

```python
# Kullanılan Modül ve Kütüphaneler
import ipywidgets as widgets
from IPython.display import display
import xml.etree.ElementTree as ET
```

**XML Oluşturma ve Yazma**

```python
# Form alanlarının oluşturulması
feTitle = widgets.Text(description="Başlık: ")
feSubject = widgets.Text(description="Konular: ")
feDescription = widgets.Text(description="Özet: ")
feCreator = widgets.Text(description="Yazar: ")
feLicense = widgets.Text(description="Lisans: ")
feLanguage = widgets.Text(description="Dil: ")
feIdentifier = widgets.Text(description="URL: ")

# Formun oluşturulması
form = widgets.VBox([feTitle, feSubject, feDescription, feCreator, feLicense, feLanguage, feIdentifier])

# Formun yayınlanması
display(form)

# Verileri XML dosyasına ekleyen fonksiyon
def add_data_to_xml(b):
  # Formdan gelen verilerin alınması
  title = feTitle.value
  subject = feSubject.value
  description = feDescription.value
  creator = feCreator.value
  license = feLicense.value
  language = feLanguage.value
  identifier = feIdentifier.value


  # XML dosyasının oluşturulması, varsa açılması
  try:
      tree = ET.parse('oermetadata.xml')
      root = tree.getroot()
  except FileNotFoundError:
      root = ET.Element('OERs')
      tree = ET.ElementTree(root)

  # Yeni bir kaynak elementi oluşturun ve verileri ekleyin
  kaynak = ET.SubElement(root, 'OER')
  ET.SubElement(kaynak, 'Title').text = title
  ET.SubElement(kaynak, 'Subject').text = subject
  ET.SubElement(kaynak, 'Description').text = description
  ET.SubElement(kaynak, 'Creator').text = creator
  ET.SubElement(kaynak, 'License').text = license
  ET.SubElement(kaynak, 'Language').text = language
  ET.SubElement(kaynak, 'Identifier').text = identifier

  # Üst verilerin XML dosyasına kaydedilmesi
  tree.write('oermetadata.xml', encoding='utf-8', xml_declaration=True)

  print("Veriler XML dosyasına eklendi!")

# Butona tıklandığında fonksiyonu çağırın
button = widgets.Button(description="Gönder")
button.on_click(add_data_to_xml)
display(button)
```

**XML Dosyasını Okuma ve Listeleme**

```python
# XML dosyasının alınması ve analiz edilmesi
tree = ET.parse('oermetadata.xml')
root = tree.getroot()

# Kaynakların okunması ve yazdırılması
for kaynak in root.findall('OER'):
    title = kaynak.find('Title').text
    subject = kaynak.find('Subject').text
    description = kaynak.find('Description').text
    creator = kaynak.find('Creator').text
    license = kaynak.find('License').text
    language = kaynak.find('Language').text
    identifier = kaynak.find('Identifier').text

    print(f"Başlık: {title}")
    print(f"Anahtar Kelimeler: {subject}")
    print(f"Özet Bilgi: {description}")
    print(f"Oluşturan: {creator}")
    print(f"Lisans: {license}")
    print(f"Dil: {language}")
    print(f"URL: {identifier}")
    print("-" * 20)  # Kaynaklar arası ayırıcı
```

**Arama**

```python
def tum_alanlarda_arama_yap(aranan_deger):
    tree = ET.parse('oermetadata.xml')
    root = tree.getroot()
    eslesen_kaynaklar = []

    for kaynak in root.findall('OER'):
        kaynak_metni = " ".join([element.text or "" for element in kaynak.iter()])
        if aranan_deger.lower() in kaynak_metni.lower():
            eslesen_kaynaklar.append(kaynak)

    return eslesen_kaynaklar

# Arama terimi için input widget'ı oluştur
arama_terimi_widget = widgets.Text(
    placeholder='Arama terimini girin',
    description='Arama:',
    disabled=False
)

# Arama butonu oluştur
arama_butonu = widgets.Button(description="Ara")

# Arama sonuçlarını görüntülemek için output widget'ı oluştur
output_widget = widgets.Output()

# Arama butonuna tıklandığında çalışacak fonksiyon
def arama_yap(b):
    aranan_deger = arama_terimi_widget.value
    sonuclar = tum_alanlarda_arama_yap(aranan_deger)

    with output_widget:
        output_widget.clear_output()  # Önceki sonuçları temizle

        if sonuclar:
            for kaynak in sonuclar:
                print(f"Başlık: {kaynak.find('Title').text}")
                print(f"URL: {kaynak.find('Identifier').text}")
                # ... diğer alanları da yazdır ...
                print("-" * 20)
        else:
            print("Eşleşen kaynak bulunamadı.")

# Arama butonuna fonksiyonu bağla
arama_butonu.on_click(arama_yap)

# Widget'ları görüntüle
display(widgets.HBox([arama_terimi_widget, arama_butonu]))
display(output_widget)
```
